```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from matplotlib import pyplot as plt
from utils import calculate_income_1000_customers
from sklearn.model_selection import GridSearchCV

TRAIN_SIZE = 0.8
np.random.seed(0)
X = pd.read_csv('../data/x_train.txt', sep=' ', header=None).to_numpy()
y = pd.read_csv('../data/y_train.txt', sep=' ', header=None).to_numpy().ravel()
```

```python
selected = [102, 103, 105]
```

```python
model = GaussianNB(var_smoothing=1)
income = []
accuracy = []
accuracy_top = []

for _ in range(10000):
    X_selected = X[:, selected]
    X_train, X_test, y_train, y_test = train_test_split(X_selected, y, train_size=TRAIN_SIZE, shuffle=True)
    model.fit(X_train, y_train)
    
    y_proba = model.predict_proba(X_test)
    y_pred = model.predict(X_test)
    
    accuracy_top.append(calculate_income_1000_customers(len(selected), y_proba=y_proba, y_true=y_test, y_pred=y_pred)[0]) 
    income.append(calculate_income_1000_customers(len(selected), y_proba=y_proba, y_true=y_test, y_pred=y_pred)[1])
    
    accuracy.append(accuracy_score(y_test, y_pred)) 
    
    
```

```python
np.mean(accuracy)
```

```python
np.mean(accuracy_top)
```

```python
np.mean(income)
```
