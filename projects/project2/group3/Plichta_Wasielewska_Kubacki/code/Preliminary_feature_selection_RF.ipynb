## Preliminary feature selection method - RF

```python
import pandas as pd
import numpy as np
import os
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import SelectFromModel
```

```python
# Define the directory containing the data files
data_directory = 'C:/Users/HP/Desktop/Advanced_ML/Projects/Project2'

# File paths
x_train_path = os.path.join(data_directory, 'x_train.txt')
y_train_path = os.path.join(data_directory, 'y_train.txt')
x_test_path = os.path.join(data_directory, 'x_test.txt')

# Load the data
X_train = pd.read_csv(x_train_path, header=None, sep='\s+')
y_train = pd.read_csv(y_train_path, header=None, sep='\s+')
X_test = pd.read_csv(x_test_path, header=None, sep='\s+')

# Assign column names to features for better clarity
feature_names = [f'feature_{i+1}' for i in range(X_train.shape[1])]
X_train.columns = feature_names
X_test.columns = feature_names
```

```python
# Display basic info about the dataset
print("Training features shape:", X_train.shape)
print("Training labels shape:", y_train.shape)
print("Test features shape:", X_test.shape)
```

```python
# Build the Random Forest classifier
rf = RandomForestClassifier(n_estimators=100)
rf.fit(X_train, y_train.values.ravel())

# Select features based on importance
selector = SelectFromModel(rf, prefit=True, max_features=10, threshold=-np.inf)
X_train_reduced = selector.transform(X_train)
X_test_reduced = selector.transform(X_test)

# Feature indices selected
selected_features = selector.get_support(indices=True)
print("Selected (10) feature indices:", selected_features)
```
